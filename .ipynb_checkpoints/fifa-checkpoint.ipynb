## Reading Data

```python
import pandas as pd
y  = pd.read_csv('data.csv')
y.head(100)
```

### Q1 - List of all Players above age 40 and Total Number of Player

```python
above_forty = y.loc[y['Age'] > 40]

count_forty = above_forty.groupby(['Age']).count().sum()
print("Total Number of Players above 40: ",count_forty.iloc[3])

print('List Of All Players above 40')
above_forty.sort_values('Age',ascending=False)
```

### Q2 - List of all Player's Names of Brazil and export them into a csv file

```python
brazil = y.loc[y['Nationality'] == 'Brazil']
bplayer = brazil.Name
bplayer
```

```python
brazil.groupby(['Nationality']).count()
print("Total Players of Brazil: ",brazil.ID.count())
```

```python
bplayer.to_csv('Brazil Players.csv',index=False) # Exporting list of all Brazil player's Name to CSV
```
