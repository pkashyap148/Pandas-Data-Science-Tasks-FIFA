## Reading Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
y  = pd.read_csv('data.csv')
y
```

### Q1 - List of all Players above age 40 and Total Number of Player

```python
above_forty = y.loc[y['Age'] > 40]

count_forty = above_forty.groupby(['Age']).count().sum()
print("Total Number of Players above 40: ",count_forty.iloc[3])

print('List Of All Players above 40')
above_forty.sort_values('Age',ascending=False)
```

```python
x = above_forty['Name']
y = above_forty.Age
plt.figure(figsize=(12,8))
plt.yticks(np.arange(0,50,2))
plt.bar(x,y)
plt.title("Age graph of Players above 40")
plt.xlabel('Player Names', fontdict={'fontsize': 12})
plt.ylabel('Ages', fontdict={'fontsize': 12})
plt.show()
```

### Q2 - List of all Player's Names of Brazil and export them into a csv file

```python
brazil = y.loc[y['Nationality'] == 'Brazil']
bplayer = brazil.Name
bplayer
```

```python
brazil.groupby(['Nationality']).count()
print("Total Players of Brazil: ",brazil.ID.count())
```

```python
bplayer.to_csv('Brazil Players.csv',index=False) # Exporting list of all Brazil player's Name to CSV
```

### Q3 - List of all players of Manchester City whose Potenital is above 90 

```python
y.head()
```

```python
y.loc[(y['Club']== 'Manchester City') & (y['Potential'] > 90)]
```

```python
low = y.loc[(y['Club']== 'Manchester City') & (y['Potential'] <= 90)].count()[0]
high = y.loc[(y['Club']== 'Manchester City') & (y['Potential'] > 90)].count()[0]

labels = ['less than 90','above 90']
plt.figure(figsize = (8,8))
plt.pie([low,high], labels = labels, autopct='%.2f %%')
plt.title('Potential of Players of Manchester City', fontdict={'fontsize': 15})
plt.show()
```

### Q-4 Name, Age, Nationality and Club Name of all Players having Overall points above 90 and Export to csv

```python
y = pd.read_csv('Data.csv')
```

```python
over = y.loc[y['Overall'] > 90]
overall = over[['Name','Age','Club','Overall']]
overall.to_csv('Overall.csv',index=False)
overall
```

```python
plt.figure(figsize = (15,5))
xa = over.Name
ya = over.Overall

plt.title("Players and their Overall", fontdict={'fontsize': 15})
plt.xlabel('\nName of Players', fontdict={'fontsize': 12})
plt.ylabel('\nOverall ', fontdict={'fontsize': 12})
plt.plot(xa,ya, 'r.--', markersize=12, label="Overall Points ")

plt.legend()
plt.show()
```
